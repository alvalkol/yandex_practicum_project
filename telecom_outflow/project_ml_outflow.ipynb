## Задача проекта

- Необходимо прогнозировать отток клиентов для оператора связи на основании персональных данных, информации о тарифах и договорах. Информация актуальна на 01.02.2020 г. Прогнозирование необходимо осуществлять без учёта даты регистрации.

<b>Описание данных</b>

Данные состоят из следующих файлов:
- `contract.csv` - иноформация о договорах;

- `personal.csv` - персональные данные клиента;

- `internet.csv` - информация об интернет-услугах;

- `phone.csv` - информация об услугах телефонии.

В файлах содержатся таблицы со следующими признаками:
- `customerID` - уникальный id клиента (есть во всех таблицах);

- `BeginDate` - дата начала пользования услугами (дата);

- `EndDate` - дата окончания пользования услугами (No, 'дата');

- `Type` - тип оплаты (Month-to-month, One year, Two year);

- `PaperlessBilling	` - безналичный расчет (Yes, No);

- `PaymentMethod` - способ оплаты (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic));

- `MonthlyCharges` - ежемесячные траты на услуги;

- `TotalCharges` - всего потрачено денег на услуги;

- `InternetService` - тип интернет-соединения (DSL, Fiber optic);

- `OnlineSecurity` - блокировка небезопасных сайтов (Yes, No);

- `OnlineBackup` - облачное хранилище файлов (Yes, No);

- `DeviceProtection` - антивирус (Yes, No);

- `TechSupport` - выделенная линия технической поддержки (Yes, No);

- `StreamingTV` - стриминговое телевидение (Yes, No);

- `StreamingMovies` - каталог фильмов (Yes, No);

- `gender` - пол (Femal, Male);

- `SeniorCitizen` - наличие пенсионного статуса по возрасту (0, 1);

- `Partner` - налиичие супруга/и (Yes, No);

- `Dependents` - наличие иждивенцев (Yes, No);

- `MultipleLines` - наличие возможности ведения параллельных линий (Yes, No);

- Для подтверждения гепотезы используем статистический метод - критерий Шапиро-Уилка для проверки нормальности распределения.

H0: Распределение нормальное

H1: Распределение не похоже на нормальное

```python
data = total_df['MonthlyCharges']

stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05

if p > alpha:
    print('Распределение нормальное (принимаем H0)')
else:
    print('Распределение не похоже на нормальное (отвергаем H0)')
```

- Для проверки гипотезы также проведем стат. тест.

```python
data = np.log(total_df['TotalCharges'])

stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p > alpha:
    print('Распределение нормальное (принимаем H0)')
else:
    print('Распределение не похоже на нормальное (отвергаем H0)')
```

## Исследовательский анализ

### Контрактные признаки

```python
total_df['Number'] = range(len(total_df))
total_df = total_df.reset_index(drop=True)
```

```python
outflow = total_df.query('OutFlow == 1')[['MonthlyCharges', 'TotalCharges', 'UsePeriod']]
no_outflow = total_df.query('OutFlow == 0')[['MonthlyCharges', 'TotalCharges', 'UsePeriod']]

fig, [ax1, ax2, ax3] = plt.subplots(nrows=3)

for key, value in {ax1: 'MonthlyCharges', ax2: 'TotalCharges', ax3: 'UsePeriod'}.items():
    key.hist(outflow[value], color='blue', label='outflow', bins=30, density=True)
    key.hist(no_outflow[value], color='yellow', label='no_outflow', bins=30, alpha=0.8, density=True)
    key.title.set_text(f'Распределение {value}')
    key.legend()
    key.grid()

fig.set_figwidth(8)
fig.set_figheight(10)
plt.show()
```

```python
display('Для оттока', outflow.describe(), 'Для остальных', no_outflow.describe())
```

- Показатели точно влияют на отток. Чем выше месячный платеж, тем выше уровень оттока. Максимальный отток в первые месяцы с максимальным общим платежом.
- Пик оттока приходиться на первые 5 месяцев использования. Основная масса клиетов уходят в течении 2х лет.
- Распределение величин платежей и периода пользования для тех, кто ушёл и нет, выглядит одинаково. Средние показатели разнятся. Месячные платежи в среднем выше для тех, кто ушёл. Общие платежи в среднем выше для тех кто продолжает контракт.

- Проверку гепотезы о равенстве средних значений `['MonthlyCharges', 'TotalCharges', 'UsePeriod']` проверим с помощью критерия Стьюдента

H0: средние значения для оттока и нет признаков `['MonthlyCharges', 'TotalCharges', 'UsePeriod']` равны


H1:  средние значения для оттока и нет признаков `['MonthlyCharges', 'TotalCharges', 'UsePeriod']` не равны

```python
for name in ['MonthlyCharges', 'TotalCharges', 'UsePeriod']:
    sample_1 = outflow[name]
    sample_2 = no_outflow[name]

    results = st.ttest_ind(sample_1,
                           sample_2, 
                           equal_var=False, )
    print('Распределение', name)
    print('p-значение:', results.pvalue)

    if results.pvalue < alpha:
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
    print()
```

- Соотношение пользователей DSL для ушедших клиентов резко отличается от оставшихся (соотношение практически один к одному). Пользователей DSL в 2 раза меньше чем Fiber.
- Получается, что при заполнении пропусков для ушедших клиентов соотношение практически не измениться. А для оставшихся может сильно повлиять на общую картину. Из графика видно, что основная часть оттока пользовалась Fiber optic

```python
def month_category(row):
    
    """
    Функция возращает категорию клиента в зависимости от его месячного платежа.
    """
    
    if  0 <= row <= 22:
        return 'base'
    elif  22 < row <= 28:
        return 'standart'
    elif  28 < row <= 75:
        return 'improved'
    else:
        return 'premium'
```

```python
total_df['MonthlyCategory'] = total_df['MonthlyCharges'].apply(month_category)
```

```python
(total_df
    .fillna('missing')
    .groupby(['OutFlow', 'MonthlyCategory', 'InternetService'])['InternetService']
    .count()
    .plot.bar(figsize=(10,6))
)
```

- Распределение интернет услуг и телефонии выгляди похожим у оттока и у оставшихся клиентов. Как и ожидалось, у категории `premium` основной вид интернета `Fiber optic`.
- Основной отток произошёл у клиентов с `Fiber optic`.
- Пропущенные значения есть только для категории `base` и `standart`. Следовательно и пропуски можно заполнить исходя из категории клиента: вид интернет-соединения `DSL`, доп.услуги - `No`

```python
total_df['InternetService'] = total_df['InternetService'].fillna('DSL')

total_df.loc[:, 'OnlineSecurity':'MultipleLines'] = (
    total_df.loc[:, 'OnlineSecurity':'MultipleLines'].fillna('No')
)
```

## Обучение различных алгоритмов

### LinearRegression & RandomForestClassifier 

```python
clfs = ((RandomForestClassifier(), {'max_features': ['auto', 'sqrt', 'log2', None],
                                    'max_depth': range(3, 15),
                                    'criterion': ['gini', 'entropy'],
                                    'min_samples_leaf': range(1, 10),
                                    'class_weight': ['balanced'],
                                    'min_samples_split': range(2, 10),
                                    }),
        (LogisticRegression(), {'penalty': ['l2'],
                                'C': [x / 10.0 for x in range(1, 50)],
                                'fit_intercept': [True, False],
                                'class_weight': ['balanced'],
                                'max_iter': [2000]
                                }))

for clf, param in clfs:
    grid = RandomizedSearchCV(estimator=clf,
                            param_distributions=param,
                            scoring='roc_auc',
                            n_iter=200,
                            n_jobs=2,
                            refit=True,
                            cv=3,
                            verbose=1,
                            error_score=0,
                            random_state=25072022)
    grid.fit(pd.get_dummies(X_train, drop_first=True), y_train)
    print('Best estimator is {} with score {} using params {}'.format(clf.__class__.__name__, 
                                                                      grid.best_score_, 
                                                                      grid.best_params_))
```

```python
lrc = LogisticRegression(class_weight='balanced',
                         random_state=25072022)
lrc.fit(pd.get_dummies(X_train, drop_first=True), y_train)

imp = pd.Series(lrc.coef_[0], 
                pd.get_dummies(X_train, drop_first=True).columns)

fig, ax = plt.subplots(figsize=(12,8))
imp.sort_values().plot.barh(ax=ax)
ax.set_title("Вес коэффициента")
ax.set_ylabel('Признаки')
fig.tight_layout()
```

```python
rfc = RandomForestClassifier(class_weight='balanced', 
                             random_state=25072022)

rfc.fit(pd.get_dummies(X_train, drop_first=True), y_train)

imp = pd.Series(rfc.feature_importances_,
                         pd.get_dummies(X_train, drop_first=True).columns)

fig, ax = plt.subplots(figsize=(12,8))
imp.sort_values().plot.barh(ax=ax)
ax.set_title("Важность признаков")
ax.set_ylabel('Важность')
fig.tight_layout()
```

- Для модели `RandomForest` метрика выше.

- Все наши созданные признаки имеют корреляции с целевым признаком, кроме `MonthlyCatagory`.

- Для обеих моделей важным при обучении оказался признак `Type`.
